```python
merged['clean_firstname'] = merged.contributor_firstname.fillna("") #took it out, but can use inplace=True to make the change permanent to the column. But that is discouraged 
```

```python
merged['clean_fullname'] = merged.clean_firstname + " " + merged.contributor_lastname
```

```python
def standardize_name(clean_fullname):
    clean_fullname = clean_fullname.strip() #strips all the whites space out of the names
    if clean_fullname == "SEAN PARKER AND AFFILIATED ENTITIES":
        return "SEAN PARKER"
    elif clean_fullname == "FUND FOR POLICY REFORM (NONPROFIT 501 (C)(4))":
        return "FUND FOR POLICY REFORM"
    else:
        return clean_fullname
    #A function should almost always have a "return." Good way to generate new columns and clean things up.
    #here we are cleaning up various names. can do multiple changes at once but gets messy if you do too many
```

```python
merged['standardized_name'] = merged.clean_fullname.apply(standardize_name)
```

```python
bigdogs = merged.groupby(
    ["standardized_name", "committee_position"]
).amount.sum().reset_index().sort_values(
    "amount", 
    ascending=False
) 
```
